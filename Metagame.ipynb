# 2017-2018 Season 1 Metagame (Modern)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def how_many(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(p=pct,v=val)
    return my_autopct
```

## The Lotus Market

```python
df = pd.read_excel('./Files/Season_1_2017-2018.xlsx', sheetname='LotusMarket', header=None, names=['Deck', 'Qty'])
```

### Show the 5 most played decks

```python
ax = df.head().plot(kind='bar', figsize=(14,6), fontsize=16, legend='', x=df['Deck'].head())
ax.set_yticks([0,1,2,3,4])
plt.show()
```

### Show total participation

```python
df.plot.pie(y='Qty', labels=df['Deck'],
            figsize=(10,10), legend=None, fontsize=16,
            autopct=how_many(df['Qty']), pctdistance=0.9)
plt.show()
```
